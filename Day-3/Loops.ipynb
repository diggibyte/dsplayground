### While loop


<li>The while loop in Python is used to iterate over a block of code as long as the test expression (condition) is true. We generally use this loop when we don't know the number of times to iterate beforehand.</li>

### Syntax:



### Flowchart of While loop:



```python
# simple while loop condition
count = 0
while(count < 3):
    print('The count is:  ', count)
    count = count + 1
print('Good bye! count is: ', count)
```

## Membership Operators

<b>in and not in</b> are membership operators in Python.

Used to find whether a value or variable is found in a sequence.

```python
# checking the values is present int list or not using membership operators
lst = [1,2,3,4]

print(5 in lst) # gives us a boolean value.
```

```python
# checking the key is present in dictionary or not 
dict = {1:'a', 2:'b'}

print('a' not in dict)

print('a' in dict)
```

```python
s = 'Python'

print('p' in s)
```

## for loop
<p> <li>A for loop is used for iterating over a sequence (that is either a list, a tuple, a dictionary, a set, or a string).</li>

<li>This is less like the for keyword in other programming languages, and works more like an iterator method as found in other object-orientated programming languages.</li>

<li>With the for loop we can execute a set of statements, once for each item in a list, tuple, set etc.</li></p>


### syntax:



### flowchart of for loop:


```python
# for loop works overs the sequence
tup = (10, 20, 30, 40)

for x in tup:
    print(x)
    
print('Finish')
```

```python
# product of elements in tuple using for loop
tup = (1,2,3,4,5)

prod = 1

for item in tup:
    prod *= item

print('Product = {}'.format(prod))
```

```python
# checking the range of values with step
for count in range(1, 10, 3):
    print('The count is:  ', count)
print('Good bye!')
```

```python
# for loop over the sequence
lst = [10,20,30,40,50,60]
for element in range(len(lst)):
    print(lst[element], end=' ')
```

```python
# iteration over the elements
print('LIST ITERATION:')
lst = [10,20,30,40,50,60]
for element in lst:
    print(element, end=' ')
print()
```

```python
print('TUPLE ITERATION:')
tup = (10,20,30,40,50,60)
for element in tup:
    print(element, end=' ')
print()
print("*"*25)

print('STRING ITERATION:')
myStr = 'Python'
for element in myStr:
    print(element, end=' ')
print()
print("*"*25)

print('DICTIONARY ITERATION:')
dict = {}
dict['xyz'] = 123
dict['abc'] = 345
for key in dict:
    print(key, dict[key])
```

### for.....else

```python
# The else block just after for loop is executed only when the loop is NOT terminated by a break statement
tup = (1,2,3,4,5)

prod = 1

for item in tup:
    prod *= item
else:
    print('You have exhausted the list')

print('Product = {}'.format(prod))
```

## Why else?

<li>while loop executes the block until a condition is satisfied. When the condition becomes false, the statement immediately after the loop is executed.</li>
<li>The else clause is only executed when your while condition becomes false. If you break out of the loop, or if an exception is raised, it wont be executed.</li>

## Nested Loops

```python
i = 0

while i < 2:
    j = 0
    while j < 3:
        print('i: {}, j:{}'.format(i, j))
        j += 1
        
    i += 1
    
print('I am outside the while loop')
```

```python
adj = ["red", "big", "tasty"]
fruits = ["apple", "banana", "cherry"]

for x in adj:
    for y in fruits:
        print(x, y)
```

### Break Keyword
<li>The purpose the break statement is to break out of a loop early. For example if the following code asks a use input a integer number x. If x is divisible by 5, the break statement is executed and this causes the exit from the loop</li>
<li>The break keyword is used to break out a for loop, or a while loop.</li>





## Break statement in for loop

```python
# break uses for early stops the loop with condition.
for count in range(10):
    if count == 5:
        break
    print('The count is:  ', count)
print('Good bye!')
```

```python
# iterating the characters in diggibyte and stopping or break the loop if it finds letters "g" or "i".  
for letter in 'diggibyte':
# break the loop as soon it sees 'g' or 'i'
    if letter == 'g' or letter == 'i':
        break
    print('Current Letter :', letter)
```

```python
# iterating over the item and stopping the loop when condition is true
lst = [1,2,3,4]

for item in lst:
    if item == 3:
        continue
    print(item)
else:
    print('Loop Ends')
```

## Break statement in while loop

```python
# here the loops stops if user enters a odd number.
while True:
    num = int(input("Enter a number to check if its even or odd: "))
    if num % 2 == 0:
        print('Congratulations, you entered an even number.')
    else:
        break

print('Wrong Number \nGAME OVER')
```

### Continue Keyword:
<li>The continue keyword is used to end the current iteration in a for loop (or a while loop), and continues to the next iteration</li>





```python
# continue skips the particular condition element and continue till the end of loop.
for count in range(10):
    if count == 5:
        continue
    print(count)
print('Good bye!')
```

```python
# Prints all numbers in list except 3.
i = 0
a = [1,2,3,4,5]

while i < len(a):
    if a[i] == 3:
        i += 1
        continue

    print('Current numbers :', a[i])
    i += 1
```
