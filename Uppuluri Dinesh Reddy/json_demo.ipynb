### Schema created while reading on json data

```
from pyspark.sql.types import StructType, StructField, StringType, DoubleType, ArrayType, IntegerType

s = StructType([
    StructField("batters", StructType([
        StructField("batter", ArrayType(StructType([
          StructField("element", StructType([
              StructField("id", IntegerType(), True),
              StructField("type", StringType(), True)
          ]))  
        ])))
        
    ])),
    StructField("id", IntegerType(),True),
    StructField("name", StringType(),True),
    StructField("ppu", DoubleType(),True),
    StructField("topping", ArrayType(StructType([
        StructField("element", StructType([
            StructField("id", IntegerType(), True),
            StructField("type", StringType(), True)
        ]))
    ]))),
    StructField("type", StringType(), True)
])
```

```
df_s = spark.createDataFrame(data = data, schema=s)
display(df_s)
```

```
df_s.printSchema()
```
